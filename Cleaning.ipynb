```python
# Imports
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, StandardScaler
```

```python
# Read in data
df = pd.read_csv('./data/IBRD_Statement_Of_Loans_-_Historical_Data.csv')
dfchina = pd.read_excel('./data/CLA_2022.xlsx')
```

```python
# Clean World Data Loans

df = df.head(100000)

df = df.drop(['Loan Number', 'Project ID', 'Project Name ', 'Cancelled Amount', 'Undisbursed Amount', 'Repaid 3rd Party', 'Due 3rd Party', 'Loans Held'], axis=1)
date_cols = ['End of Period', 'First Repayment Date', 'Last Repayment Date', 'Agreement Signing Date', 'Board Approval Date', 'Effective Date (Most Recent)', 'Closed Date (Most Recent)', 'Last Disbursement Date']
df[date_cols] = df[date_cols].apply(pd.to_datetime)
cat_cols = ['Region', 'Country Code', 'Borrower', 'Guarantor Country Code', 'Guarantor', 'Loan Type', 'Loan Status', 'Currency of Commitment']

enc = OneHotEncoder()
for col in cat_cols:
    if col in df.columns:
        encoded = enc.fit_transform(df[[col]])
        df = df.drop(col, axis=1)
        cols = [f'{col}_{cat}' for cat in enc.categories_[0]]
        df[cols] = pd.DataFrame(encoded.toarray(), columns=cols)

num_cols = ['Interest Rate', 'Original Principal Amount', 'Disbursed Amount', 'Repaid to IBRD', 'Due to IBRD', 'Exchange Adjustment', "Borrower's Obligation"]
scaler = StandardScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

date_cols = ['End of Period', 'First Repayment Date', 'Last Repayment Date', 'Agreement Signing Date', 'Board Approval Date', 'Effective Date (Most Recent)', 'Closed Date (Most Recent)', 'Last Disbursement Date']
df = df.drop(date_cols, axis=1)

cat_cols = ['Country', 'Loan Type', 'Interest Rate Type']
enc = OneHotEncoder()
for col in cat_cols:
    if col in df.columns:
        encoded = enc.fit_transform(df[[col]])
        df = df.drop(col, axis=1)
        cols = [f'{col}_{cat}' for cat in enc.categories_[0]]
        df[cols] = pd.DataFrame(encoded.toarray(), columns=cols)

df = df.dropna()
```

```python
# Clean China Data set

dfchina.columns = ['Loan_ID', 'Year', 'Country', 'Lender', 'Lender_Type', 'USD_M', 'Purpose', 'Sector']

dfchina.drop(['Loan_ID'], axis=1, inplace=True)

dfchina['Purpose'].fillna('Unknown', inplace=True)

dfchina['USD_M'] = dfchina['USD_M'].map('{:.2f}'.format)

dfchina.drop_duplicates(inplace=True)

dfchina.reset_index(drop=True, inplace=True)

print(dfchina.head())
```
